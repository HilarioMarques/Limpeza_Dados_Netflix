#### Tratando os valores nulos da coluna diretor

<p>Temos algumas formas de fazer isso:
    <ul>
        <li>Podemos passar o valor de "não informado" para esses valores nulos</li>
        <li>Podemos também procurar e passar os nomes dos diretores de cada obra que está com o valor NAN, entretanto, isso exige um trabalho muito longo e cansativo, <strong>você deve se perguntar se vale o tempo e se é realmente necessário fazer isso?</strong></li>
    </ul>
</p>

```python
base['director'].isnull().sum() #somando a quantidade de valores nulos na coluna director, observa que se eu fosse fazer uma pesquisa para substituir os valores nulos por cada diretor correto, eu iria fazer isso 2634 vezes, então isso poderia ser um atrasdo no projeto, se por acaso o meu chefe ou supervisor chegasse e falasse que seria importante para o projeto ter esses dados corretos, aí sim eu deveria pesquisar e passar os valores exatos
```

```python
base.loc[base['director'].isnull(), 'director'] = "uninformed" #o 'loc' nos permite acessar os valores nulos em determinada coluna e atribuir um valor a eles, nesse caso estou atribuido 'uninformed' (não informado), para esses valores
```

```python
base['director'].isnull().sum() #verificando se ainda existe algum valor nulo
```

```python
base['director'].nunique() #verificando a cardinalidade e descobrindo que há 4529 valores diferentes na coluna de diretores
```

```python
base['director'].value_counts().head(10) #verificando os 10 autores que mais aparecem, observe que uninformed é o que mais aparece, ou seja, a base na coluna director está bem incompleta 
```

#### Tratando a coluna de country(países) e rating(classificação de maturidades)

- Os países nulos, se encaixam no mesmo caso dos diretores, com pesquisa na internet seria fácil encontrar onde essas obras foram produzidas

```python
base.loc[base['country'].isnull(), 'country'] = 'uninformed'
```

```python
#coluna de rating
base['rating'].isnull().sum()
```

```python
#Essa linha de código verifica quais os valores presentes nessa coluna e quantas vezes eles aparecem, veja que os 3 ultimos, são dados de duração, decidir passar esses valores para NAN, dessa forma, posso ajustar de maneira correta
base['rating'].value_counts()
```

```python
#percebi que todos esses dados de duração que estão em rating, são de um mesmo diretor
base[base['director'] == 'Louis C.K.']
```

```python
#vamos fazer essa substituição, usando os títulos
import numpy as np

base.loc[base['title'] == 'Louis C.K. 2017', 'duration'] = '74 min'
base.loc[base['title'] == 'Louis C.K. 2017', 'rating'] = np.nan
base.loc[base['title'] == 'Louis C.K. 2017']
```

```python
base.loc[base['title'] == 'Louis C.K.: Hilarious', 'duration'] = '84 min'
base.loc[base['title'] == 'Louis C.K.: Hilarious', 'rating'] = np.nan
base.loc[base['title'] == 'Louis C.K.: Hilarious']
```

```python
base.loc[base['title'] == 'Louis C.K.: Live at the Comedy Store', 'duration'] = '66 min'
base.loc[base['title'] == 'Louis C.K.: Live at the Comedy Store', 'rating'] = np.nan
base.loc[base['title'] == 'Louis C.K.: Live at the Comedy Store']
```

```python
base['rating'].value_counts()
```

```python
base['rating'].isnull().sum() #tem apenas 7 dados vazios de rating
```

```python
#Essa coluna de rating está muito específica, por exemplo: G e TV-G significa que é liberado para o público geral, então poderíamos unir por classes, óbvio que esssa união depende do projeto

base.loc[(base['rating'] == 'TV-G') | (base['rating'] == 'G') | (base['rating'] == 'TV-Y7') | (base['rating'] == 'TV-Y7-FV') | (base['rating'] == 'TV-Y') | (base['rating'] == 'TV-PG') | (base['rating'] == 'PG'), 'rating'] = 'R-C' #recomendado para crianças

base.loc[(base['rating'] == 'PG-13') | (base['rating'] == 'TV-14'), 'rating'] = 'R-A' #recomendado para adolescentes

base.loc[(base['rating'] == 'R') | (base['rating'] == 'TV-MA') | (base['rating'] == 'NC-17'), 'rating'] = 'R+17' #recomendado para maiores

#A sigla NR significa não classificado e a sigla UR, aparentemente são filmes que tem algumas cenas bem eróticas e que segundo algumas comunidades na internet, são proibidos em alguns cinemas grandes nos EUA
base['rating'].value_counts()
```

```python
base[base['rating'].isnull()]
```

- Como tinha apenas 7 dados vazios na coluna de rating, eu decidir pesquisar o rating e cada obra, a atribuição dos valores foi usando o loc e puxando os dados pelo o show_id, as fontes desses resultados foram o própio site da Netflix e o iMDB

```python
base.loc[base['show_id'] == 's5542', 'rating'] = 'R+17'
base.loc[base['show_id'] == 's5814', 'rating'] = 'R+17'
base.loc[base['show_id'] == 's5795', 'rating'] = 'R+17'
base.loc[base['show_id'] == 's5990', 'rating'] = 'R-A'
base.loc[base['show_id'] == 's6828', 'rating'] = 'R-A'
base.loc[base['show_id'] == 's7313', 'rating'] = 'R-C'
base.loc[base['show_id'] == 's7538', 'rating'] = 'R-A'
```
